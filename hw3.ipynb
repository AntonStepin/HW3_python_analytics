## Задача 3
### Условие 1

Найдите среднюю стоимость домов в зависимости от количества спален

Отсортируйте от меньшей стоимости к большей

```python
group_by_bed = df.groupby('bedrooms').agg({'price': 'mean'}).sort_values('price')
group_by_bed
```

### Условие 2

Найдите минимальную, среднюю и максимальную стоимости домов в зависимости от состояния дома

```python
price_cond = df.groupby('condition').agg({'price': ['min', 'mean', 'max']})
price_cond
```

### Условие 3
Постройте таблицу с подсчетом количества домов в данных в зависимости от вида на набережную и оценкой вида

```python
df.pivot_table(index="waterfront", columns="view", values="price", aggfunc='count').fillna(0)
#pd.options.display.float_format = '{:,.0f}'.format
```

### Условие 4

Каких домов в зависимости от этажности и количества спален больше?

```python
df.pivot_table(index="floors", columns="bedrooms", values="price", aggfunc='count', fill_value=0)
```

```python
df.groupby(['floors', 'bedrooms']).agg({'id': 'count'}).sort_values('id', ascending=False).head(1)
```
